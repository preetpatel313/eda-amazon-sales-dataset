Changing Data Types of Columns from object to float

```python
df['actual_price'] =df['actual_price'].str.replace('₹','')
df['actual_price']=df['actual_price'].str.replace(',','')
df['actual_price']= df['actual_price'].astype(float)
```

```python
df['discounted_price']= df['discounted_price'].str.replace('₹','')
df['discounted_price']  = df['discounted_price'].str.replace(',','')
df['discounted_price']= df['discounted_price'].astype(float)
```

```python
df.info()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.replace('%','')
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
df['rating'].value_counts()
```

```python
df.query('rating ==  "|"')
```

I got this product rating on Amazon by searching the provided product_id on their official website (amazon.in)¶
The rating is 3.9. So, I am going to give the item rating a 3.9 as well.

```python
df['rating'] = df['rating'].str.replace('|','3.9').astype(float)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',','')
df['rating_count'] =df['rating_count'].astype(float)
```

Heatmap
# Plot correlations between variables

```python
categorical_cols = [
    'product_id',
    'category',
    'review_id',
    'review_content',
    'product_name',
    'user_name',
    'about_product',
    'user_id',
    'review_title',
    'img_link',
    'product_link'
]

for col in categorical_cols:
    df[col] = df[col].astype('category').cat.codes
```

```python

cols = [
    'discounted_price',
    'actual_price',
    'discount_percentage',
    'rating',
    'rating_count'
]

corr = df[cols].corr()

plt.figure(figsize=(8,5))

sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title("Sales Related Correlation")
plt.show()
```

The analysis shows a very strong positive correlation between actual price and discounted price (0.96). Discount percentage has a weak negative relationship with prices, while ratings and rating counts show very little correlation with pricing variables.

 Conduct t-test to compare rating between two categories

```python
import scipy.stats as stats

#import scipy.stats as stats

electronics = df[df['category'] == 2]['rating']
clothing = df[df['category'] == 1]['rating']

t_statistic, p_value = stats.ttest_ind(electronics, clothing)

print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Significant difference between categories")
else:
    print("No significant difference between categories")
```

Q9: What are the Top 5 categories based with highest ratings?

```python
# Top 5 categories with highest average ratings
top_categories = (
    copydf.groupby('category')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

print(top_categories)
```

```python

# Top 5 categories
top_categories = (
    copydf.groupby('category')['rating']
    .mean()
    .sort_values(ascending=False)
    .head(5)
)

plt.figure(figsize=(10,5))

ax = sns.barplot(
    x=top_categories.values,
    y=top_categories.index,
    hue=top_categories.index,
    palette='viridis',
    legend=False
)

# Add values
for i, v in enumerate(top_categories.values):
    ax.text(v + 0.01, i, f"{v:.2f}", va='center')

plt.title("Top 5 Categories with Highest Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Category")

plt.show()
```

Answer 9:The top-rated categories were Tablets, Network Adapters, Camera Accessories, Memory Components, and Streaming Devices. Among them, the Tablets category had the highest average rating of 4.6.

```python
Q10: What is the sentiment distribution of product reviews?
```

TextBlob was used to perform sentiment analysis on customer reviews and classify them as positive, negative, or neutral based on polarity scores.

✅ Sentiment Polarity Range
+1 → Positive review
0 → Neutral review
-1 → Negative review

Example:

“Excellent product” → Positive
“Bad quality” → Negative

```python
!pip install textblob
```

```python

from textblob import TextBlob

# Calculate sentiment polarity
copydf['sentiment'] = copydf['review_content'].apply(
    lambda text: TextBlob(str(text)).sentiment.polarity
)

# Categorize sentiment
copydf['sentiment_label'] = copydf['sentiment'].apply(
    lambda x: 'Positive' if x > 0 else 'Negative' if x < 0 else 'Neutral'
)

# Count sentiment types
sentiment_counts = copydf['sentiment_label'].value_counts()

print(sentiment_counts)
```

Answer 10:Sentiment analysis was performed on product reviews using TextBlob. The results showed that the majority of reviews were positive, with 1438 positive reviews, indicating high customer satisfaction. Only 26 reviews were negative and 1 review was neutral, suggesting that customers generally had a positive experience with the products.
